```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV
df = pd.read_csv('sales.csv')
```

```python
# Convert the 'sales' and 'expenditure' columns to integers
df['sales'] = df['sales'].astype(int)
df['expenditure'] = df['expenditure'].astype(int)
```

```python
#General Analysis of the Data
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
# Descriptive statistics
print(df.describe().round(2))
```

```python
# group by the “Rep” column and sum up the values in the columns.
df.groupby(["sales"]).sum().sort_values("expenditure", ascending=False)
```

```python
# Calculate total sales and average sales
total_sales = df['sales'].sum()
average_sales = df['sales'].mean()

print(f'Total Sales: {total_sales}')
print(f'Average Sales: {average_sales:.2f}')
```

```python
# Extracting data for visualization
months = df['month']
sales = df['sales']
expenditures = df['expenditure']
```

```python
# Plotting using bar chart
plt.figure(figsize=(12, 6))
plt.bar(months, sales, label='sales', color='blue', alpha=0.7)
plt.bar(months, expenditures, label='expenditure', color='red', alpha=0.5)  # Alpha for transparency
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Monthly Sales and Expenditure')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Plotting using line chart
plt.figure(figsize=(12, 6))
plt.plot(months, sales, label='Sales', color='blue', marker='o')
plt.plot(months, expenditures, label='Expenditure', color='red', marker='x')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Trend of Monthly Sales and Expenditure')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
